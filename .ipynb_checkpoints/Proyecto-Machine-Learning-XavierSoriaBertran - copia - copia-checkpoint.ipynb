```python
# cargamos librerías necesarias
import numpy  as np  
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt # para dibujar
%matplotlib inline
```

# Cargar datos

```python
airbnb = pd.read_csv("./data/airbnb-listings-extract.csv", sep=';', decimal='.')   
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
pd.set_option('display.float_format', '{:.1f}'.format)
```

```python
airbnb.shape
```

```python
#Nos quedamos con los apartamentos de la ciudad de Madrid
airbnb= airbnb[airbnb.City == 'Madrid']
```

```python
airbnb.shape
```

Podemos ver (sin entrar a mirar los datos) que hay una serie de variables que no nos aportan ningún tipo de información, para la variable objetivo del proyecto, por lo que procedemos a eliminarlas, antes de dividir en train y test, y realizar el análisis como tal. Podemos considerar que estas variables son: 

ID - Listing Url - Scrape ID - Last Scraped - Name - Summary - Space - Description - Experiences Offered - Notes - Transit - Access - Interaction - House Rules	- Thumbnail Url	- Medium Url - Picture Url - XL Picture Url	- Host ID - Host URL - Host Name - Host Since - Host Location - Host About - Host Thumbnail Url	- Host Picture Url - Host Neighbourhood	- Host Listings Count - Host Total Listings Count - Host Verifications - Country Code - Country - Calendar Updated - Calendar last Scraped - First Review - Last Review	- License - Jurisdiction Names - Calculated host listings count	- Features

40 columnas en total

```python
#Eliminamos las columnas descritar en la celda anterior 
airbnb = airbnb.drop(['ID','Listing Url', 'Scrape ID', 'Last Scraped', 'Name', 'Summary', 'Space', 'Description', 'Experiences Offered', 'Notes', 'Transit', 'Access', 'Interaction', 'House Rules', 'Thumbnail Url',  'Medium Url',  'Picture Url',  'XL Picture Url', 'Host ID',  'Host URL', 'Host Name',  'Host Since', 'Host Location', 'Host About', 'Host Thumbnail Url', 'Host Picture Url', 'Host Neighbourhood', 'Host Listings Count', 'Host Total Listings Count', 'Host Verifications', 'Country Code', 'Country', 'Calendar Updated', 'Calendar last Scraped', 'First Review', 'Last Review', 'License', 'Jurisdiction Names', 'Calculated host listings count', 'Features'], axis=1)
```

```python
airbnb.shape
```

# Análisis exploratorio y Preprocesamiento

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(airbnb, test_size=0.2, shuffle=True, random_state=0)

print(f'Dimensiones del dataset de training: {train.shape}')
print(f'Dimensiones del dataset de test: {test.shape}')

# Guardamos
train.to_csv('./train.csv', sep=';', decimal='.', index=False)
test.to_csv('./test.csv', sep=';', decimal='.', index=False)
```

```python
df_train_airbnb = pd.read_csv('./train.csv', sep=';', decimal='.')
```

```python
round(df_train_airbnb.isnull().mean()*100,2).sort_values(ascending=False)
```

Eliminamos las columnas que tienen > 50% de valores ausentes, ya que como se ha comentado, no tiene sentido hacer una imputación, debido a que los valores estarían "desvirtuados".


```python
df_train_airbnb = df_train_airbnb.drop(['Host Acceptance Rate', 'Has Availability', 'Square Feet', 'Monthly Price', 'Weekly Price', 'Security Deposit'], axis=1)   
```

```python
round(df_train_airbnb.isnull().mean()*100,2).sort_values(ascending=False)
```

Imputamos los valores nulos con la moda, a excepción del servicio de limpieza que lo imputamos con 0

```python
for column in df_train_airbnb:
    if column == 'Cleaning Fee':
        df_train_airbnb[column].fillna(0, inplace=True)
    else:
        df_train_airbnb[column].fillna(df_train_airbnb[column].mode()[0], inplace=True)
```

```python
#round(df_train_airbnb.isnull().mean()*100,2)
```

Mirando una muestra de los datos, vemos que aún nos quedan variables que no aportan información relevante (en su mayoría categóricas) o bien no son necesarias (se explica). Por lo que procedemos a eliminarlas. Estas variables son:

Neighborhood Overview, Street, Neighbourhood (no es necesaria ya que si vemos los porcentajes de null y los porcentajes de null de Neighbourhood Cleansed, vemos que en Neighbourhood son muy elevados, y en Neighbourhood Cleansed son 0. Asimismo, viendo la muestra de datos vemos que Neighbourhood Cleansed completa prácticamente al 100% a la variable Neighbourhood), City, State, Market, Smart Location, Amenities, Maximum Nights, Geolocation

```python
df_train_airbnb = df_train_airbnb.drop(['Neighborhood Overview', 'Street', 'Neighbourhood', 'City', 'State', 'Market', 'Smart Location', 'Amenities', 'Maximum Nights', 'Geolocation'], axis=1)  
```

Veamos que variables categóricas nos quedan después de una primera limpieza y las transformamos

```python
df_train_airbnb.dtypes
```

Eliminamos ZipCode porque es una variable categórica cuando debería ser numérica.

```python
df_train_airbnb = df_train_airbnb.drop(['Zipcode'], axis=1)  
```

```python
from sklearn.preprocessing import LabelEncoder

le_Host_Response_Time = LabelEncoder()
le_Host_Response_Time.fit(df_train_airbnb['Host Response Time'])
df_train_airbnb['Host Response Time'] = le_Host_Response_Time.transform(df_train_airbnb['Host Response Time'])

le_Neighbourhood_Cleansed = LabelEncoder()
le_Neighbourhood_Cleansed.fit(df_train_airbnb['Neighbourhood Cleansed'])
df_train_airbnb['Neighbourhood Cleansed'] = le_Neighbourhood_Cleansed.transform(df_train_airbnb['Neighbourhood Cleansed'])

le_Neighbourhood_Group_Cleansed = LabelEncoder()
le_Neighbourhood_Group_Cleansed.fit(df_train_airbnb['Neighbourhood Group Cleansed'])
df_train_airbnb['Neighbourhood Group Cleansed'] = le_Neighbourhood_Group_Cleansed.transform(df_train_airbnb['Neighbourhood Group Cleansed'])

le_Property_Type = LabelEncoder()
le_Property_Type.fit(df_train_airbnb['Property Type'])
df_train_airbnb['Property Type'] = le_Property_Type.transform(df_train_airbnb['Property Type'])

le_Room_Type = LabelEncoder()
le_Room_Type.fit(df_train_airbnb['Room Type'])
df_train_airbnb['Room Type'] = le_Room_Type.transform(df_train_airbnb['Room Type'])

le_Bed_Type = LabelEncoder()
le_Bed_Type.fit(df_train_airbnb['Bed Type'])
df_train_airbnb['Bed Type'] = le_Bed_Type.transform(df_train_airbnb['Bed Type'])

le_Cancellation_Policy = LabelEncoder()
le_Cancellation_Policy.fit(df_train_airbnb['Cancellation Policy'])
df_train_airbnb['Cancellation Policy'] = le_Cancellation_Policy.transform(df_train_airbnb['Cancellation Policy'])
```

```python
df_train_airbnb_filtered = df_train_airbnb_filtered[['Price', 'Host Response Time', 'Host Response Rate', 'Neighbourhood Cleansed',
       'Neighbourhood Group Cleansed', 'Latitude', 'Longitude',
       'Property Type', 'Room Type', 'Accommodates', 'Bathrooms', 'Bedrooms',
       'Bed Type', 'Cleaning Fee', 'Guests Included',
       'Extra People', 'Minimum Nights', 'Availability 30', 'Availability 365',
       'Number of Reviews', 'Review Scores Rating', 'Review Scores Accuracy',
       'Review Scores Cleanliness', 'Review Scores Checkin',
       'Review Scores Communication', 'Review Scores Location',
       'Cancellation Policy', 'Reviews per Month']]
```

```python
from sklearn import preprocessing

data = df_train_airbnb_filtered.values
y_train = data[:,0:1]     # nos quedamos con la 1ª columna, price
X_train = data[:,1:]      # nos quedamos con el resto
feature_names = df_train_airbnb_filtered.columns[1:]

# Escalamos (con los datos de train)
scaler = preprocessing.StandardScaler().fit(X_train)
XtrainScaled = scaler.transform(X_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import tree


maxDepth = range(1,15)
param_grid = {'max_depth': maxDepth }
grid = GridSearchCV(DecisionTreeRegressor(random_state=0), param_grid=param_grid, cv = 10)
grid.fit(XtrainScaled, y_train)
print("best mean cross-validation score: {:.3f}".format(grid.best_score_))
print("best parameters: {}".format(grid.best_params_))

scores = np.array(grid.cv_results_['mean_test_score'])
plt.plot(maxDepth,scores,'-o')
plt.xlabel('max_depth',fontsize=16)
plt.ylabel('10-Fold MSE')
plt.show()
```

```python
from sklearn.model_selection import KFold

N,Nfeatures = XtrainScaled.shape

rf = RandomForestRegressor(max_depth=maxDepthOptimo,n_estimators=200,max_features='sqrt')
kf  = KFold(n_splits=10, shuffle = True, random_state=1)
cv_error = []
cv_std = []

for nfeatures in range(Nfeatures,0,-1):
        
    error_i = []
    
    for idxTrain, idxVal in kf.split(XtrainScaled):
        
        Xt = XtrainScaled[idxTrain,:]
        yt = y_train[idxTrain]
        Xv = XtrainScaled[idxVal,:]
        yv = y_train[idxVal]
        
        rf.fit(Xt,yt)
        
        ranking = rf.feature_importances_
        indices = np.argsort(ranking)[::-1] 
    
        selected = indices[0:(Nfeatures-nfeatures+1)]
        
        Xs = Xt[:,selected]
        
        rf.fit(Xs,yt)
        error = (1.0-rf.score(Xv[:,selected],yv))     
        error_i.append(error) 
    
    cv_error.append(np.mean(error_i))
    cv_std.append(np.std(error_i))
    
    print('# features ' + str(len(selected)) + ' error ' + str(np.mean(error_i)) + ' +/- ' + str(np.std(error_i)))
```

```python
plt.plot(range(1,Nfeatures+1,1),cv_error,'-o')
plt.errorbar(range(1,Nfeatures+1,1), cv_error, yerr=cv_std, fmt='o')
plt.xlabel('# features')
plt.ylabel('CV error')
plt.show()
```

LASSO

```python
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.linear_model import Lasso

tscv = TimeSeriesSplit(n_splits=8)

alpha_vector = np.logspace(-4,4,20)
param_grid = {'alpha': alpha_vector}

grid = GridSearchCV(Lasso(), param_grid=param_grid, cv = tscv.split(XtrainScaled)).fit(XtrainScaled, y_train)
```

```python
print("best mean cross-validation score: {:.3f}".format(-1*grid.best_score_))
print("best parameters: {}".format(grid.best_params_))

scores = grid.cv_results_['mean_test_score']
std_scores = grid.cv_results_['std_test_score']
plt.errorbar(np.log10(alpha_vector),scores,yerr=std_scores, fmt='o-',ecolor='g')
plt.xlabel('log(alpha)',fontsize=16)
plt.ylabel('CV MSE')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

alpha_optimo = grid.best_params_['alpha']
lasso = Lasso(alpha = alpha_optimo).fit(XtrainScaled,y_train)

ytrainLasso = lasso.predict(XtrainScaled)
ytestLasso  = lasso.predict(XtestScaled)

mseTrainModelLasso = mean_squared_error(y_train,ytrainLasso)
mseTestModelLasso = mean_squared_error(y_test,ytestLasso)

print('MSE Modelo Lasso (train): %0.3g' % mseTrainModelLasso)
print('MSE Modelo Lasso (test) : %0.3g' % mseTestModelLasso)

print('RMSE Modelo Lasso (train): %0.3g' % np.sqrt(mseTrainModelLasso))
print('RMSE Modelo Lasso (test) : %0.3g' % np.sqrt(mseTestModelLasso))
```
